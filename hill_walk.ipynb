```python
def clash(brd):
    board=[]
    for i in range(n):
        board.append((brd[i],i))

    cnt=0
    for i in range(n-1):
        val=board[i]
        for j in range(i+1,n):
            val1=board[j]
            if(abs(val[0]-val1[0])==abs(val[1]-val1[1])):
                cnt+=1
            if(val[0]==val1[0]):
                cnt+=1
    
    return cnt
```

```python
def succ_find(pos):
    positions.clear()
    lst=[]
    for i in range(n):
        for j in range(n):
            if(pos[i]!=j):
                lst=pos.copy()
                lst[i]=j
                lst=list(lst)
                positions.append(lst)
    
    for p in positions:
        print(p,clash(p))
    print("\n\n\n")
    return positions
```
